# Simple graph visualization

RDFLIB includes simple graph rendering capability, convenient for smaller graphs. 

Ntoe that attempting to render large graphs (e.g. the schema.org ontology) will 
consume a lot of memory and will likely fail.

```python
import rdflib
import rdflib.tools.rdf2dot
import graphviz
import io

g = rdflib.ConjunctiveGraph()
g.parse(
    "data/incomplete-eg-01.jsonld", 
    format="json-ld", 
    publicID="https://science-on-schema.org/examples/data/")

fp = io.StringIO()
rdflib.tools.rdf2dot.rdf2dot(g, fp)
graphviz.Source( fp.getvalue() )
```
